```python
from sklearn.datasets import load_iris
```

```python
fifaData=pd.read_csv("/content/Top_650_FIFA.csv")
```

```python
columnas_x=["pace","shooting","passing","dribbling","defending","physic","attacking","skill","movement","power","mentality","overall","age","height_cm","weight_kg","weak_foot","skill_moves"]
```

```python
fifaData_date_x=fifaData[columnas_x]
```

```python
fifaData_data_y=fifaData.age
```

```python
fifaData_shortnames=fifaData.short_name.tolist()
```

```python
x_train,x_test,y_train,y_test=train_test_split(fifaData_date_x,fifaData_data_y,test_size=0.3 ,random_state=1)#generamos 4 variables sobre nuesto set de datos 
```

```python
classifier=DecisionTreeClassifier(max_depth=6).fit(x_train,y_train) #pasamos nuestro arbol de decisiones a nuetsra variable classifier.
```

```python
datos_dot=export_graphviz(classifier, out_file=None, filled=True,rounded=True, feature_names=columnas_x, class_names=fifaData_shortnames)
```

```python
grafica=pydotplus.graph_from_dot_data(datos_dot)
```

```python
imagen=grafica.write_png("Arbol2.png")
```
