```python
import pandas as pd
from sklearn.model_selection import train_test_split#para dividir nuestra data entre entrenamiento y testing
from sklearn.tree import DecisionTreeClassifier,export_graphviz#  importar el clasificador de arbol de decisiones y  graficar
import pydotplus
from IPython.display import Image
from google.colab import files
from sklearn.datasets import load_iris
```

```python

uploaded=files.upload()
```

```python
columnas_x=["Energy","Length."]
```

```python
topData=pd.read_csv("/content/top50_nuevos - top50.csv")
```

```python
topData_date_x=topData[columnas_x]
```

```python
topData_data_y=topData.Energy
```

```python
topData_shortnames=topData.Name.tolist()
```

```python
x_train,x_test,y_train,y_test=train_test_split(topData_date_x,topData_data_y,test_size=0.3 ,random_state=1)#generamos 4 variables sobre nuesto set de datos 
```

```python
classifier=DecisionTreeClassifier(max_depth=6).fit(x_train,y_train) #pasamos nuestro arbol de decisiones a nuetsra variable classifier.
```

```python
datos_dot=export_graphviz(classifier, out_file=None, filled=True,rounded=True, feature_names=columnas_x, class_names=topData_shortnames)
```

```python
grafica=pydotplus.graph_from_dot_data(datos_dot)
```

```python
imagen=grafica.write_png("Arbol3.png")
```
